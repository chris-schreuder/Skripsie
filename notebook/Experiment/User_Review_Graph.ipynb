```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dicts={'Score out of 5': [4.75, 3.5, 4.5, 4.25], 'Review aspects': ['Visual appeal', 'User Experience', 'Accurate Simulation', 'Overall Score']}
df = pd.DataFrame.from_dict(dicts)
plt.figure(1, [20, 8])
df.plot.bar(x='Review aspects', y='Score out of 5', rot=10, width=0.5)
for index, value in enumerate(dicts['Score out of 5']):
    plt.text(index-0.1, value-0.2,
             str(value))
```
